### Calculating elasticities and diversion ratios.

```python
%matplotlib inline

import pyblp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pyblp.options.digits = 2
pyblp.options.verbose = False
```

```python
product_data = pd.read_csv(pyblp.data.NEVO_PRODUCTS_LOCATION)
agent_data = pd.read_csv(pyblp.data.NEVO_AGENTS_LOCATION)
product_formulations = (
   pyblp.Formulation('0 + prices', absorb='C(product_ids)'),
   pyblp.Formulation('1 + prices + sugar + mushy')
)
agent_formulation = pyblp.Formulation('0 + income + income_squared + age + child')
problem = pyblp.Problem(product_formulations, product_data, agent_formulation, agent_data)
problem
```

```python
product_data
```

```python
initial_sigma = np.diag([0.3302, 2.4526, 0.0163, 0.2441])
initial_pi = [
  [ 5.4819,  0,      0.2037,  0     ],
  [15.8935, -1.2000, 0,       2.6342],
  [-0.2506,  0,      0.0511,  0     ],
  [ 1.2650,  0,     -0.8091,  0     ]
]
results = problem.solve(
    initial_sigma,
    initial_pi,
    optimization=pyblp.Optimization('l-bfgs-b', {'gtol': 1e-6}),
    method='1s'
)
results
```

```python
elasticities = results.compute_elasticities()
diversions = results.compute_diversion_ratios()
```

```python
import matplotlib.pyplot as plt

# Define masks for the selected markets
single_market_20Q1 = product_data['market_ids'] == 'C20Q1'
single_market_20Q2 = product_data['market_ids'] == 'C20Q2'
single_market_30Q1 = product_data['market_ids'] == 'C30Q1'
single_market_30Q2 = product_data['market_ids'] == 'C30Q2'

# Get the elasticities for the selected markets
elasticities_20Q1 = elasticities[single_market_20Q1]
elasticities_20Q2 = elasticities[single_market_20Q2]
elasticities_30Q1 = elasticities[single_market_30Q1]
elasticities_30Q2 = elasticities[single_market_30Q2]

# Find the global min and max values for the color scale across all markets
vmin = min(elasticities_20Q1.min(), elasticities_20Q2.min(), elasticities_30Q1.min(), elasticities_30Q2.min())
vmax = max(elasticities_20Q1.max(), elasticities_20Q2.max(), elasticities_30Q1.max(), elasticities_30Q2.max())

# Create a figure with two subplots side by side for City 20
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the elasticities for City 20 Q1 Market
cax1 = axes[0].matshow(elasticities_20Q1, cmap='viridis', vmin=vmin, vmax=vmax)
fig.colorbar(cax1, ax=axes[0])
axes[0].set_title('C20 Q1 Market')

# Plot the elasticities for City 20 Q2 Market
cax2 = axes[1].matshow(elasticities_20Q2, cmap='viridis', vmin=vmin, vmax=vmax)
fig.colorbar(cax2, ax=axes[1])
axes[1].set_title('C20 Q2 Market')

# Add a main title to the figure
fig.suptitle('Demand Elasticities by Product in City 20', fontsize=16)

# Display the plots
plt.show()

# Create a figure with two subplots side by side for City 30
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the elasticities for City 30 Q1 Market
cax1 = axes[0].matshow(elasticities_30Q1, cmap='viridis', vmin=vmin, vmax=vmax)
fig.colorbar(cax1, ax=axes[0])
axes[0].set_title('C30 Q1 Market')

# Plot the elasticities for City 30 Q2 Market
cax2 = axes[1].matshow(elasticities_30Q2, cmap='viridis', vmin=vmin, vmax=vmax)
fig.colorbar(cax2, ax=axes[1])
axes[1].set_title('C30 Q2 Market')

# Add a main title to the figure
fig.suptitle('Demand Elasticities by Product in City 30', fontsize=16)

# Display the plots
plt.show()
```
